```python
import json
import sys

# Make sure the script's directory is importable
sys.path.insert(0, '/content')

from hf_iqr_v3_subset_builder import build_subset, sha256_of_obj

# Load V2 master dataset
with open('HF_IQR_Master_Dataset_v2.json', 'r', encoding='utf-8') as f:
    master = json.load(f)

# Build V3 subset
v3 = build_subset(master, seed=80)

# Write the output
with open('HF_IQR_V3_Dataset_80.json', 'w', encoding='utf-8') as f:
    json.dump(v3, f, indent=2, ensure_ascii=False)

# Report
subset_hash = sha256_of_obj(v3['dataset'])
print(f"V3 subset built.")
print(f"  seed:             {v3['seed']}")
print(f"  total questions:  {v3['total_questions']}")
print(f"  categories:       {v3['total_categories']}")
print(f"  V3 dataset hash:  {subset_hash}")
print()
print("Per-category selection (record with the seed):")
for cat, ids in v3['selection_log'].items():
    print(f"  {cat:20s} ({len(ids)}): {', '.join(ids)}")
```
